```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data_f = pd.read_csv('flipkart_com-ecommerce_sample.csv')
data_f.head()
```

```python
data_f.shape
```

```python
data_f.info()
```

```python
data_f.columns
```

```python
data_f.drop(data_f.columns.difference(['product_name', 'retail_price', 'discounted_price','pid']), 1, inplace=True)
```

```python
data_a = pd.read_csv('amz_com-ecommerce_sample.csv', encoding='latin1')
data_a.head()
```

```python
data_a.shape
```

```python
data_a.info()
```

```python
data_a.columns
```

```python
data_a.drop(data_a.columns.difference(['product_name', 'retail_price', 'discounted_price','pid']), 1, inplace=True)
```

```python
final= pd.merge(data_f, data_a, on="pid")
final
```

```python
final.drop('pid',axis=1,inplace=True)
```

```python
final= final.rename(columns={'product_name_x': 'Flipkart Product name',
                    'retail_price_x': 'Flipkart Retail price',
                     'discounted_price_x': 'Flipkart Discounted price',
                    'product_name_y': 'Amazon Product name',
                    'retail_price_y': 'Amazon Retail price',
                     'discounted_price_y': 'Amazon Discounted price',})
```

```python
final.head(20)
```

```python
final.to_csv('matched_data.csv')
```
